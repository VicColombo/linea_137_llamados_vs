### Pasar las columnas  a categorical

#### Más normalizaciones
- ver todas las cat: llamado_provincia, llamante_vinculo, hecho_lugar, victima_nacionalidad, victima_vinculo_agresor

- en todo el dataset Ns/Nc = NS/NC
- llamado provincia id pasar a integer
- llamante_edad Sin dato pasar a N/A
- llamante género: Masculino = Masculino Trans = Transgénero
- llamante vinculo Vecino = Vecina/o ' Madre' = 'Madre'
- caso judicializado 'NS/NS' = 'NS/NC' 'Sin datos' = 'NS/NC'
- victima_a_resguardo No = NO
- victima_genero Trans = Transgénero
- victima_vinculo_agresor 'Pareja de la vícitma' = 'Pareja de la víctima' 'Pareja ' = 'Pareja de la víctima'  Ex pareja = Ex pareja de la víctima
- hecho_lugar Otro(Especificar en observaciones) = Otro 
- llamado_provincia: 'Ciudad Autónoma de Buenos Aires' = 'CABA'
- llamado_provincia: 'Santa Fé' = 'Santa Fe'


```python
llamados_entero.llamante_genero.replace({' Masculino':'Masculino', 'Trans':'Transgénero'}, inplace = True)
llamados_entero.llamante_vinculo.replace({'Vecino': 'Vecina/o', ' Madre':'Madre'}, inplace = True)
llamados_entero.victima_a_resguardo.replace({'No': 'NO'}, inplace = True)
llamados_entero.victima_genero.replace({'Trans': 'Transgénero'}, inplace = True)
llamados_entero.victima_vinculo_agresor.replace({'Pareja de la vícitma': 'Pareja de la víctima', 'Pareja ': 'Pareja de la víctima','Pareja': 'Pareja de la víctima', 'Ex pareja': 'Ex pareja de la víctima'}, inplace = True)
llamados_entero.hecho_lugar.replace({'Otro(Especificar en observaciones)': 'Otro' }, inplace = True)
llamados_entero.caso_judicializado.replace({'NS/NS': 'NS/NC', 'Sin datos':'NS/NC'}, inplace = True)
llamados_entero.llamado_provincia.replace({'Ciudad Autónoma de Buenos Aires': 'CABA'}, inplace = True)
llamados_entero.llamado_provincia.replace({'Santa Fé': 'Santa Fe'}, inplace = True)
llamados_entero.replace('Ns/Nc', 'NS/NC', inplace=True)
```
